```python
df_selected = df[[
    "City",
    "State",
    "Name",
    "Google review rating",
    "Number of google review in lakhs"
]]
```

```python
df_selected = df_selected.rename(columns={
    "City": "city",
    "State": "state",
    "Name": "place",
    "Google review rating": "rating",
    "Number of google review in lakhs": "popularity"
})
```

```python
df_selected.head()
```

```python
df_selected.isnull().sum()
```

```python
df_selected = df_selected.dropna(subset=["rating", "popularity"])
```

```python
df_selected["rating"] = df_selected["rating"].astype(float)
df_selected["popularity"] = df_selected["popularity"].astype(float)
```

```python
df_selected.info()
```

```python
def filter_by_source_city(df, source_city):
    filtered_df = df[df["city"].str.lower() != source_city.lower()].copy()
    return filtered_df
source_city = "Delhi"
filtered_df = filter_by_source_city(df_selected, source_city)
filtered_df.head()
```

```python
filtered_df["rating_score"] = (
    filtered_df["rating"] / filtered_df["rating"].max()
)
filtered_df["popularity_score"] = (
    filtered_df["popularity"] / filtered_df["popularity"].max()
)
filtered_df.head()
```

```python
filtered_df["final_score"] = (
    0.6 * filtered_df["rating_score"] +
    0.4 * filtered_df["popularity_score"]
)

filtered_df.sort_values(
    by="final_score",
    ascending=False
).head(10)
```

```python
def weekend_getaway_recommender(df, source_city, top_n=5):
    filtered_df = df[
        df["city"].str.lower() != source_city.lower()
    ].copy()
    filtered_df["rating_score"] = (
        filtered_df["rating"] / filtered_df["rating"].max()
    )
    filtered_df["popularity_score"] = (
        filtered_df["popularity"] / filtered_df["popularity"].max()
    )
    filtered_df["final_score"] = (
        0.6 * filtered_df["rating_score"] +
        0.4 * filtered_df["popularity_score"]
    )

    return filtered_df.sort_values(
        by="final_score",
        ascending=False
    ).head(top_n)[[
        "city", "state", "place", "rating", "popularity", "final_score"
    ]]
```

```python
weekend_getaway_recommender(df_selected, "Delhi", 5)
```
